AS WE CAN SEE ABOVE THERE ARE FEW VALUES IN AGE WHICH ARE IN NAGATIVE LETS TAKE CARE OF THEM

```python
# Count the number of customers with negative ages
negative_age_count = (df_cus['age'] < 0).sum()
negative_age_count
```

```python
# Replace negative ages with their absolute values
df_cus['age'] = df_cus['age'].abs()
```

```python
df_cus.head(10)
```

THERE WERE ONLY 4 NAGATIVE VALUES AND WE HAVE TURNED THEM WITH ABSALUTE VALUES

LETS CREATE A NEW COLUMN NAMED (total_sale_amount) USING QUANTITY SOLD x UNIT PRICE

```python
# There were no column named unit price in the sales dataframe so we created one with sale_amount and quantity_sold
df_sal['unit_price'] = df_sal['sale_amount'] / df_sal['quantity_sold'].replace(0, 1)
```

```python
# Calculate total sale amount by multiplying quantity_sold and unit_price
df_sal['total_sale_amount'] = df_sal['quantity_sold'] * df_sal['unit_price']
```
